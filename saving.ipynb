```python
import pandas as pd

# Replace with your file name if different
df = pd.read_csv('personal_finance_tracker_dataset.csv')

# Show first 5 rows
print(df.head(2))
```

## Data Preprocessing

```python
print(df.info())
print(df.isnull().sum())
```

## Handle Missing Values

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

## Dropping Unnecssary Columns

```python
df = df.drop(columns=['user_id', 'date', 'actual_savings'], errors='ignore')
```

## Encode Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

# Define categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

le = LabelEncoder()

for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

## Split Features and Target

```python
X = df.drop(columns=['savings_goal_met'])
y = df['savings_goal_met']
```

## Feature Scaling and Final Output

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()   # ✅ create object first
X_scaled = scaler.fit_transform(X)

print("Shape of X:", X_scaled.shape)
print("Shape of y:", y.shape)
```

## Model training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
import numpy as np

# ---- Step 1: Train model (run once before input) ----
from sklearn.ensemble import RandomForestClassifier

# Select best 7 features
features = [
    'monthly_income',
    'monthly_expense_total',
    'savings_rate',
    'debt_to_income_ratio',
    'credit_score',
    'investment_amount',
    'financial_stress_level'
]

X = df[features]
y = df['savings_goal_met']

# Scale
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train model
model = RandomForestClassifier()
model.fit(X_scaled, y)

# ---- Step 2: Take user input ----
print("Enter user financial details:")

monthly_income = float(input("Monthly Income: "))
monthly_expense_total = float(input("Monthly Expenses: "))
savings_rate = float(input("Savings Rate (%): "))
debt_to_income_ratio = float(input("Debt-to-Income Ratio: "))
credit_score = float(input("Credit Score: "))
investment_amount = float(input("Investment Amount: "))
financial_stress_level = float(input("Financial Stress Level (1-10): "))

# ---- Step 3: Convert to array ----
user_data = np.array([[ 
    monthly_income,
    monthly_expense_total,
    savings_rate,
    debt_to_income_ratio,
    credit_score,
    investment_amount,
    financial_stress_level
]])

# ---- Step 4: Scale input ----
user_scaled = scaler.transform(user_data)

# ---- Step 5: Predict ----
prediction = model.predict(user_scaled)

# ---- Step 6: Output result ----
if prediction[0] == 1:
    print("✅ User is likely to MEET savings goal")
else:
    print("❌ User is NOT likely to meet savings goal")
```
